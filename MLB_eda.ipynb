# MLB exploratory data analysis (EDA)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MLB_height_weight.csv')
```

```python
df.dropna(inplace=True)
```

```python
pos = df.groupby('Position')
```

```python
np.polyfit(df['Height(inches)'], df['Weight(pounds)'],1)
```

```python
%matplotlib notebook

xpts = np.linspace(66,84)

for n, gp in pos:
    x, y = gp['Height(inches)'].values, gp['Weight(pounds)'].values
    plt.plot(x,y, 'k.', alpha=0.1)
    p = np.polyfit(x,y,1)
    plt.plot(xpts, p[0]*xpts + p[1],'-',label=n)
    
plt.legend()
```
